```python
import requests
```

```python
headers={
        'User-Agent':'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.89 Safari/537.36 SLBrowser/7.0.0.12151 SLBChan/25'
            }
url='https://mp.weixin.qq.com/s/VGbv5vi8BxHzA62lLHiwgg'
file_name='用了2周，终于把Python网站开发库大全整理完了'
```

```python
res=requests.get(url=url,headers=headers).text
```

```python
res
```

```python
from lxml import etree
```

```python
tree=etree.HTML(res)
```

```python
content_list=tree.xpath('//*[@id="page-content"]//text()')
```

```python
content_list
```

```python
str1=' '.join(content_list)
```

```python
str1
```

```python
str1=str1.replace('\n','').replace(' ','')
```

```python
str1=str1.replace('。','。\n').replace('！','！\n').replace('？','？\n').replace('；','；\n').replace('：','：\n')
print(str1)
```

```python
with open(file_name+'.txt','w',encoding='utf-8') as f:
    f.write(str1)
```
